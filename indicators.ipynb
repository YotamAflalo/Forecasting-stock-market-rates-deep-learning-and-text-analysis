# with indicators

```python
import math
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler 
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dropout
```

```python

stock_data= df
close_prices = stock_data['Open']
values = close_prices.values

training_data_len = math.ceil(len(values)* 0.85)

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(stock_data)
train_data = scaled_data[0: training_data_len, :]
#close_prices = scaled_data['Open']
#values = close_prices.values

x_train = []
y_train = []

for i in range(10, len(train_data)):
    x_train.append(train_data[i-10:i, :])
    y_train.append(train_data[i, 0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], x_train.shape[2]))
test_data = scaled_data[training_data_len-10: , : ]
x_test = []
y_test = []

for i in range(10, len(test_data)):
    x_test.append(test_data[i-10:i, :])
    y_test.append(test_data[i, 0])
y_test = np.array(y_test)
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], x_test.shape[2]))

model = keras.Sequential()
model.add(layers.LSTM(100, return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))

model.add(layers.LSTM(100, return_sequences=False))
model.add(layers.Dense(25))
model.add(layers.Dense(1))
model.summary()
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, batch_size= 2, epochs=5)
```

```python
predictions = model.predict(x_test)

prediction_copies= np.repeat(predictions, len(list(df.columns)), axis=-1)
prediction_copies.shape
predictions = scaler.inverse_transform(prediction_copies)[:,0]
rmse = np.sqrt(np.mean(predictions - y_test)**2)
rmse
```

```python
data = stock_data.filter(['Open'])
train = data[:training_data_len]
validation = data[training_data_len:]
validation['Predictions'] = predictions
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date')
plt.ylabel('Close Price USD ($)')
plt.plot(train)
plt.plot(validation[['Open', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

# indicators and directions

```python
stock_data= df

stock_data['lag close'] = stock_data['Close'].shift(1)
stock_data['Open_cahnge'] = stock_data['Open'] - stock_data['lag close']
stock_data =stock_data.dropna()
stock_data['morning_rash'] = np.where(stock_data['Open_cahnge']>0,1,0)
```

```python

close_prices = stock_data['morning_rash']
values = close_prices.values
training_data_len = math.ceil(len(values)* 0.8)

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(stock_data)
train_data = scaled_data[0: training_data_len, :]
#close_prices = scaled_data['Open']
#values = close_prices.values

x_train = []
y_train = []

for i in range(12, len(train_data)):
    x_train.append(train_data[i-12:i, :])
    y_train.append(train_data[i, -1])
    
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], x_train.shape[2]))


test_data = scaled_data[training_data_len-60: , : ]
x_test = []
y_test = []

for i in range(12, len(test_data)):
    x_test.append(test_data[i-12:i, :])
    y_test.append(test_data[i, -1])
y_test = np.array(y_test)
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], x_test.shape[2]))


model = keras.Sequential()
model.add(layers.LSTM(100, return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(Dropout(0.2))
model.add(layers.LSTM(100, return_sequences=False))
model.add(Dropout(0.2))
model.add(layers.Dense(25))
#model.add(layers.Dense(1,activation='sigmoid'))
model.add(layers.Dense(1,activation='hard_sigmoid'))

model.summary()
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, batch_size= 16, epochs=6, validation_split=0.1, verbose=1)
```

```python
loss_and_metrics = model.evaluate(x_test, y_test)
print('Loss = ',loss_and_metrics[0])
print('Accuracy = ',loss_and_metrics[1])
```
